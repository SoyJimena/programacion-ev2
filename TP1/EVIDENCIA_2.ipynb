```python
import random
import datetime
import time
import pandas as pd

#EDITAR NOMBRE DEL EVENTO
def editar_nombre_evento():
    validacion_folio_sala = int(input("Ingrese el folio asignado de la sala: "))
    if validacion_folio_sala in diccionario_salas.keys():
        nombre_nuevo = input("Ingrese nombre nuevo del evento: ")
        diccionario_salas[validacion_folio_sala] = nombre_nuevo
        print ("El nombre del evento fue modificado")
    else:
        print ("El folio no esta registrado. Intentelo de nuevo.")



# REGISTRO DE RESERVACION
def registrar_evento():
    global diccionario_eventos
    validacion_clave_cliente = int(input("Ingrese la clave del cliente: "))
    if validacion_clave_cliente in diccionario_clientes:
        nombre_evento = input("Ingrese el nombre del evento: ")
        numero_sala = int(input("Ingrese numero de la sala: "))
        turno = int(input("Indique el turno que desea entre [\n 1. Matutino. \n 2.Vespertino. \n 3.Nocturno] : "))
        if turno != "1" or turno != "2" or turno != "3":
            fecha_actual = datetime.date.today()
            fecha_evento_str = input ("Indique la fecha del evento en el formato dd/mm/aaaa: ")
            fecha_evento = datetime.datetime.strptime(fecha_evento_str, "%d/%m/%Y").date()
            if fecha_evento.day <= fecha_actual.day +2:
                print ("No se pudo realizar la reservacion ya que debio hacerlo con dos dias de anticipacion como minimo, lo lamentamos.")
            else:
                print ("Su reservacion se a realizado.")
            
            diccionario_eventos[numero_sala] =  nombre_evento, turno, fecha_evento
        
        else:
            print ("Debe de elegir una opcion de las anteriores.")
         
    else:
        print ("Es necesario ser cliente para hacer la reservacion.")
        
        
diccionario_eventos = {}



#REPOSITORIO 
def repositorio():

  diccionario_repositorio = pd.DataFrame(diccionario_eventos)
  diccionario_repositorio.index = list(diccionario_eventos.keys())
  print(diccionario_repositorio)


def registro_sala ():
    global diccionario_salas
    numero_sala = int(input("Ingrese numero de la sala : "))
    cupo_sala = int(input("Indique el cupo de la sala: "))
    folio_sala = max(list(diccionario_salas.keys()),default=999+1)
    if cupo_sala <= 0:
        print("Debe de haber gente en la sala.")
    else:
        print (f"El folio de su sala reservada es el: {folio_sala} ")


    diccionario_salas[folio_sala] = numero_sala, cupo_sala 

diccionario_salas = {}




def agregar_cliente():
  global diccionario_clientes
  nombre_cliente = input("Ingrese el nombre: ")
  if nombre_cliente.isspace() or len(nombre_cliente) <= 0:
      print ("Es obligatorio ingresar el nombre.")
  else:
      clave_cliente = max(list(diccionario_clientes.keys()),default=9999+1)
      print ("La clave del cliente asignado es el: ", clave_cliente)
      
      diccionario_clientes[clave_cliente] = nombre_cliente

diccionario_clientes = {}




while True:
  print("\n ** MENU PRINCIPAL **")
  print ("**********************")
  print ("1- Reservaciones")
  print ("2- Reportes")
  print ("3- Registrar sala")
  print ("4- Registrar cliente")
  print ("5- Salir")

  respuesta = input("Indique la opcion deseada: ")

  try:
    respuesta_int = int(respuesta)
  except ValueError:
    print ("* La respuesta no es valida*")
  except Exception:
    print("Se a presentado una excepcion: ")
    print(Exception)

   # RESERVACIONES
  if respuesta_int == 1:
      while True:
          print("\n ** SUBMENU RESERVACIONES **")
          print ("1- Registrar nueva reservacion")
          print ("2- Modificar descripcion de una reservacion")
          print ("3- Consulta disponibilidad de salas para una fecha")
          print ("4- Regresar al menu principal")
          
          respuesta_submenu = int(input("Indique la opcion deseada: "))
          
          try:
              respuesta_submenu_int = int(respuesta_submenu)
          except ValueError:
              print ("* La respuesta no es valida*")
          except Exception:
              print("Se a presentado una excepcion: ")
              print(Exception)
            
          if respuesta_submenu == 1:
              registrar_evento()
            
          elif respuesta_submenu == 2:
              editar_nombre_evento()
            
          elif respuesta_submenu == 3:
              print ("Consultar disponibilidad de sala")
            
          elif respuesta_submenu == 4:
              
              break
            
          else:
              print ("Ingrese una de las respuestas existentes en el menu.")


   # CAMBIAR EL NOMBRE DEL EVENTO
  elif respuesta_int == 2:
      while True:
          print("\n ** SUBMENU REPORTES **")
          print ("1- Reporte en pantalla de reservaciones para una fecha")
          print ("2- Exportar reporte tabular en Excel")
          print ("3- Regresar al menu principal")
          
          respuesta_reportes = int(input("Indique la opcion deseada: "))
          
          try:
              respuesta_reportes_int = int(respuesta_reportes)
          except ValueError:
              print ("* La respuesta no es valida*")
          except Exception:
              print("Se a presentado una excepcion: ")
              print(Exception)
            
          if respuesta_reportes == 1:
              print ("Esta opcion no se puede realizar en estos momentos.")
            
          elif respuesta_reportes == 2:
              print ("Esta opcion no se puede realizar en estos momentos.")
            
          elif respuesta_reportes == 3:
              
              break
            
          else:
              print ("Ingrese una de las respuestas existentes en el menu.")


   #REGISTRAR LA SALA
  elif respuesta_int == 3:
      registro_sala()
    
    
    # REGISTRO CLIENTES
  elif respuesta_int == 4:
      agregar_cliente()


  elif respuesta_int == 5:

    break
  else:
    print ("Su respuesta no corresponde a ninguna de las opciones")

```
